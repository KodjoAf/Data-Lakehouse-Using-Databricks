# Transformation

```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY cp.start_date, cp.product_number) AS product_key,
    cp.product_id,
    cp.product_number,
    cp.product_name,
    cp.category_id,
    epc.category,
    epc.subcategory,
    epc.maintenance_flag,
    cp.product_line,
    cp.start_date
FROM silver.crm_products cp
LEFT JOIN silver.erp_product_category epc ON epc.category_id = cp.category_id
"""
df = spark.sql(query)
#df.display()
```

```python
df.limit(5).display()
```

# Writing Gold Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_products")
```
