# Imports and bronze table reading

```python
import pyspark.sql.functions as func
from pyspark.sql.types import StringType, DateType
from pyspark.sql.functions import col, trim
from pyspark.sql.window import Window
```

```python
df = spark.table("workspace.bronze.crm_prd_info")
df.display()
```

# Transformations

### Trimming

```python
for field in df.schema.fields:
    if isinstance(field.dataType, StringType):
        df = df.withColumn(field.name, trim(col(field.name)))
     
```

### Product Key Parsing

```python
df = df.withColumn("cat_id", func.regexp_replace(func.substring(col("prd_key"), 1, 5), "-", "_"))
df = df.withColumn("prd_key", func.substring(col("prd_key"), 7, func.length(col("prd_key"))))
```

### Cost clean Up

```python
df = df.withColumn("prd_cost", func.coalesce(col("prd_cost"), func.lit(0)))    
```

### Product Line Normalization

```python
df = (
    df
    .withColumn(
        "prd_line",
        func.when(func.upper(col("prd_line")) == "M", "Mountain")
         .when(func.upper(col("prd_line")) == "R", "Road")
         .when(func.upper(col("prd_line")) == "S", "Other Sales")
         .when(func.upper(col("prd_line")) == "T", "Touring")
         .otherwise("n/a")
    )
)
```

### Date Casting

```python
df = df.withColumn("prd_start_dt", col("prd_start_dt").cast(DateType()))
```

### Renaming Columns

```python
RENAME_MAP = {
    "prd_id": "product_id",
    "cat_id": "category_id",
    "prd_key": "product_number",
    "prd_nm": "product_name",
    "prd_cost": "product_cost",
    "prd_line": "product_line",
    "prd_start_dt": "start_date",
    "prd_end_dt": "end_date"
}
for old_name, new_name in RENAME_MAP.items():
    df = df.withColumnRenamed(old_name, new_name)
     
```

### Sanity Checks

```python
df.limit(10).display()
```

# Writing Silver Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.silver.crm_products")
```
