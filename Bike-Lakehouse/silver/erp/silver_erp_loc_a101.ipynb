# Imports and bronze table reading

```python
import pyspark.sql.functions as func
from pyspark.sql.types import StringType, DateType
from pyspark.sql.functions import col, trim
```

```python
df = spark.table("workspace.bronze.erp_loc_a101")
```

# Transformations

### Trimming

```python
for field in df.schema.fields:
    if isinstance(field.dataType, StringType):
        df = df.withColumn(field.name, trim(col(field.name)))
```

### Customer ID Cleanup

```python
df = df.withColumn("cid", func.regexp_replace(col("cid"), "-", ""))
```

### Country Normalization

```python
df = df.withColumn(
    "cntry",
    func.when(col("cntry") == "DE", "Germany")
     .when(col("cntry").isin("US", "USA"), "United States")
     .when((col("cntry") == "") | col("cntry").isNull(), "n/a")
     .otherwise(col("cntry"))
)
```

### Renaming Columns

```python
RENAME_MAP = {
    "cid": "customer_number",
    "cntry": "country"
}
for old_name, new_name in RENAME_MAP.items():
    df = df.withColumnRenamed(old_name, new_name)
```

### Sanity checks

```python
df.limit(10).display()
```

# Writing Silver Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.silver.erp_customer_location")
```
